```python
def load_data():

    import kagglehub
    from kagglehub import KaggleDatasetAdapter
    file_path = "athlete_events.csv"


    df = kagglehub.load_dataset(KaggleDatasetAdapter.PANDAS,"heesoo37/120-years-of-olympic-history-athletes-and-results",file_path,)

    df['Medal'] = df['Medal'].fillna('No Medal')
    df['Age'] = round(df['Age'].fillna(df['Age'].mean()))
    df['Height'] = round(df['Height'].fillna(df['Height'].mean()))
    df['Weight'] = round(df['Weight'].fillna(df['Weight'].mean()))


    dtype_map = {'ID':'int64', 'Name':'string', 'Sex':'category', 'Age':'float64', 'Height':'float64','Weight':'float64', 'Team':'string', 'NOC':'string', 'Games':'string',
                 'Year':'string', 'Season':'string', 'City':'string', 'Sport':'string', 'Event':'string','Medal':'category'}
    df = df.astype(dtype_map)


    return df
```

```python
s = load_data()
```

```python
df_india = s[(s['Team']=='India')].copy()
df_india = df_india.reset_index()
df_india = df_india.drop(columns=['index'])
```

```python
df_indiaMedal=df_india[['Year','Sport','Medal']]
```

```python
df_indiaMedal['Count'] = 1
```

```python
df_indiaMedal=df_indiaMedal.pivot_table(index = 'Year',columns='Medal',values='Count',aggfunc='sum',fill_value=0)
```
